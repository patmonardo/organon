## Kant review linkage: logical division and inference forms

### 1) Concept of logical division (Kant)
A concept contains a manifold under itself: unity by agreement and multiplicity by difference.
Logical division determines the concept with respect to all possible cases under it, insofar as cases are opposed/distinct.
- Higher concept: **divided concept** (*divisus*)
- Lower concepts: **members of division** (*membra dividentia*)

Working bridge to this notebook: division is not optional taxonomy; it is the formal condition for handling manifold determinacy without collapsing distinctions.

### 2) Matter and form of inference of reason
- **Matter**: antecedent propositions (premises).
- **Form**: conclusion as consequentia.

Compiler-relevant reading: matter = claim inventory; form = licensed transition rule producing an output judgment.

### 3) Division by relation: categorical, hypothetical, disjunctive
Kant grounds exactly three forms from three conditions of objective unity of consciousness:
1. **Inherence of marks in a subject** -> categorical relation
2. **Dependence of one cognition on another as ground** -> hypothetical relation
3. **Combination of parts in a whole (logical division)** -> disjunctive relation

So the triad of inference forms is generated by relation-conditions of unity, not by rhetorical style.

### 4) Why this is directly relevant to our U-P-S / Being-Nothing-Becoming work
Provisional alignment for review:
- **Universal / Being as such**: stable subject-side frame (inherence condition).
- **Particular / Nothing-as-Shine**: conditional dependence and selective determination (ground/dependence condition).
- **Singular / Becoming as such**: resolved determination through partitioned alternatives and this-instance binding (whole/parts and decision closure).

Hypothesis: Hegel’s first movement in the Logic can be read as transforming these Kantian relation-forms from fixed logical molds into immanent self-moving determinations.
