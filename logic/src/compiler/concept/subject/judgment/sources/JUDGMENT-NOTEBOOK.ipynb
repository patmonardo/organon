## Dialectical Dyad Progression (Judgment A–D)

### Core chain

- Guiding progression: `Genus↔Species` dyads are progressively transformed into `Subject↔Predicate` dyads.
- The transformation is not a replacement but an immanent redetermination through the copula.
- This sequence is dialectical and dynamic (jewel-like), not a rigid static taxonomy.

### Stage specification

1. `A. Judgment of Existence`
- Dyad form: immediate singular ↔ abstract universal.
- Copula form: immediate abstract `is`.
- Limitation: contradiction between immediacy and universality.

2. `B. Judgment of Reflection`
- Dyad form: singular/particular/allness articulations within reflective mediation.
- Copula form: increasingly relational connection (subsumption logic).
- Result: objective universality emerges and reverses earlier subsumption relation.

3. `C. Judgment of Necessity`
- Dyad form (explicit): genus ↔ species (categorical), then conditioned term ↔ term (hypothetical), then genus ↔ total disjunctive species field.
- Copula form: necessity as inner-then-posited connection.
- Result: conceptive disjunction where universal identity and negative exclusion are unified.

4. `D. Judgment of the Concept`
- Dyad form: subject (singular fact with constitution) ↔ predicate (conceptive measure: good/right/true, etc.).
- Copula form: from external assertion to accomplished, contentful connection.
- Result: replete copula yields syllogistic transition.

## Dialectical Cube Spec (3x3x3) for KG Modeling

Date: 2026-02-21

Goal: define the **Dialectical Cube** as a 3x3x3 genera-structure for precise Knowledge Graph modeling, reusable for both `Transition` and `TruthOf` relations.

### Cube intuition

- 3 root threads (horizontal genera): `Being`, `Essence`, `Concept`
- 3 moments per thread (intra-thread phases)
- 3 recursive levels (meta-level strata of the same form)

This yields a 3x3x3 lattice of homologous logical positions.

### Canonical coordinate model

Represent each cube node by coordinate:

`C = (thread, moment, level)`

where

- `thread ∈ {Being, Essence, Concept}`
- `moment ∈ {1,2,3}`
- `level ∈ {1,2,3}`

Optional stable id pattern:

`cube:<thread>:m<moment>:l<level>`

Example: `cube:Being:m2:l1`

### Relation semantics over the cube

1. **Transition relation** (dynamic)

`Transition(C_i -> C_j)` is valid when at least one cube adjacency rule holds:

- `horizontal`: same level, same thread, adjacent moment
- `vertical`: same level, same moment, adjacent thread (synchronization)
- `recursive`: same thread and moment, adjacent level

2. **TruthOf relation** (grounding/verification)

`TruthOf(X, C_k)` means proposition/process `X` is true in virtue of cube position `C_k` and its induced neighborhood constraints.

Operationally:

- `Transition` captures movement constraints.
- `TruthOf` captures validity constraints.
- Both read from one shared cube topology.

### Genera claim

The Dialectical Cube functions as genus-of-genera: each coordinate is not merely a point but a typed logical place that distributes isomorphically across threads and levels.

### Implementation note for Relations phase

When we formalize in workbook/KG:

- keep textual claims source-anchored first,
- then project each accepted claim onto one or more cube coordinates,
- then derive `Transition` and `TruthOf` edges from the same coordinate logic.

This keeps transitions and truth-relations unified under a single matrix grammar.
