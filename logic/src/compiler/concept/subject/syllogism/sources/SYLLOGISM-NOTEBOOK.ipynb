## Foundation, Ground, and Foundering (working meditation)

Date: 2026-02-21

In this phase of reading, **Foundering** names the movement where immediate determinacy cannot sustain itself and passes into reflected determinacy. This is not mere collapse; it is a consequential transition.

- Determinate Being -> Determinations of Reflection -> Judgment as Determinate Concept.
- Foundering appears first as loss of immediacy, but in truth it is a **transition into Ground**.
- As Consequence (`consequentia`), Ground is not externally added; it is the truth produced by the movement itself.

A working distinction for our compiler thought:

- **Foundation**: what is posited as supporting basis within a given sphere of discourse.
- **Ground**: the immanent reason that emerges through mediation and justifies the relation of determinations.

Inherence gives a second articulation of this movement:

- Positive -> Negative -> Infinite as the immanent arc of foundering.
- Thus foundering is not only privative; it is the route by which foundation is shown as finite and Ground is disclosed as truth.

### Three relation-spheres of words (for modeling)

We can model one triadic relation-form across different spheres, while preserving sphere-specific semantics:

1. **Inherence sphere** (subjective-immanent): predicates inhere in a substrate and negate toward infinity.
2. **Consequence sphere** (inferential): terms are linked through mediated necessity (`therefore`).
3. **Categorical sphere** (objective-world): determinate concept dominates as world-structure, where relation is not optional but constitutive.
