# Ground REVIEW Workbook (Notebook)

Purpose: review quality and authority compliance of Ground Part A/B/C workbooks without adding new inferential claims.

## Contract

- REVIEW is non-authoritative and audit-focused.
- Authority remains source text + Part A/B/C workbook artifacts.
- REVIEW must not invent claims or alter claim IDs.
- INFERENCE workbook remains separate and grounded by approved claims only.

## Review Protocol (Recursive Passes)

### Pass 1 — Structural Integrity
- Sectioning follows source sequence exactly.
- Entry IDs, claim IDs, and relation IDs are stable.
- Required fields exist: Span, Summary, KeyPoints, Claims, Relations, Review outcome.

### Pass 2 — Evidence Integrity
- Every claim has line-anchored evidence.
- Evidence spans match sentence boundaries (no mid-sentence clipping).
- Confidence levels are proportionate to evidence strength.

### Pass 3 — Causal/Logical Integrity
- Summary names the section's generative cause.
- KeyPoints track necessary transitions, not commentary.
- transitions_to relations are causally justified by closing claims.

## REVIEW Record Template

Use one record per reviewed entry.

```yaml
reviewRecord:
  timestamp: 2026-02-20
  reviewer: <name>
  workbookPart: <A|B|C>
  entryId: <e.g., grd-abs-a-001>
  pass: <1|2|3>
  status: <ok|needs_revision|review_pending>
  checks:
    structure:
      sectionOrder: <ok|fail>
      schemaCompleteness: <ok|fail>
      idStability: <ok|fail>
    evidence:
      lineAnchorsPresent: <ok|fail>
      sentenceBoundaryIntegrity: <ok|fail>
      claimToEvidenceFit: <ok|fail>
    causality:
      summaryStatesGenerativeCause: <ok|fail>
      keyPointsAreNecessaryTransitions: <ok|fail>
      transitionNecessity: <ok|fail>
  findings:
    strengths:
      - <bullet>
    issues:
      - <bullet>
  requestedChanges:
    - field: <Summary|KeyPoints|Claims|Relations|Span>
      action: <tighten|split|merge|re-anchor|clarify>
      rationale: <short reason>
  openQuestions:
    - <question>
```

## Example REVIEW Record (Ground Part A)

```yaml
reviewRecord:
  timestamp: 2026-02-20
  reviewer: pat
  workbookPart: A
  entryId: grd-abs-a-003
  pass: 3
  status: needs_revision
  checks:
    structure:
      sectionOrder: ok
      schemaCompleteness: ok
      idStability: ok
    evidence:
      lineAnchorsPresent: ok
      sentenceBoundaryIntegrity: ok
      claimToEvidenceFit: ok
    causality:
      summaryStatesGenerativeCause: ok
      keyPointsAreNecessaryTransitions: fail
      transitionNecessity: ok
  findings:
    strengths:
      - c1-c3 preserve internal progression from content to determinate ground.
    issues:
      - k4 reads descriptive; should state why negated opposition is required for c3.
  requestedChanges:
    - field: KeyPoints
      action: clarify
      rationale: express causal necessity for transition to twofold determinateness
  openQuestions:
    - Should k4 explicitly name the sublation mechanism as transition driver?
```

## Handoff Rule to INFERENCE

A claim is eligible for INFERENCE only if latest REVIEW status is `ok` for:
- `lineAnchorsPresent`
- `claimToEvidenceFit`
- `keyPointsAreNecessaryTransitions`

If any fails, keep INFERENCE unchanged and return to workbook revision.
