- `being/`: `quality/`, `quantity/`, `measure/`
- `essence/`: `reflection/`, `appearance/`, `actuality/`

Each directory contains highly structured, granular conceptual nodes, capturing Hegel's dialectical transitions.

### Bootstrapping Strategy

Instead of translating Yoga Sūtras directly into Fichte/Hegel ad hoc, we invert the process. We use the `@reality/hegel-logic` KG as the **Universal Schema** for the entire `@logic/compiler`.

1. **Core Coordinate System**: The 1-3-9 structure (e.g., `essence/reflection/ground`) becomes the absolute directory structure for all comparative philosophy within the Organon.
2. **Cross-Mapping**: Future translations of Yoga, Sāṃkhya, and Abhidharma will be explicitly assigned a strict Hegelian coordinate (e.g., mapping *Sādhana-pāda* into the specific dialectic of *Measure* or *Subjectivity*).
3. **Graph Generation**: By structuring the markdown workbooks with standardized tags based on the `hegel-logic` namespace, we can parse these MD files to auto-generate Cypher queries for the master Neo4j graph.

This methodology transforms the repository from a collection of translations into a unified, programmable Idea-Compiler.

## Phase 4 Expansion: The Doctrine of Being (Samāpatti)

**Objective:** Escalate the initial baseline analysis of Yoga Sūtras I.41-51 into the foundational triadic structure of the Doctrine of Being, matching the Fichtean *Science of Knowing* (Lecture 15) and Hegel's *Science of Logic* (Quality, Quantity, Measure).

### The Triads of Being
We successfully generated 11 distinct Yoga Sūtra mappings across 36 workbooks (SANSKRIT, FICHTE, HEGEL, IDEA) tracing the descent of consciousness from pure transparency to the threshold of Essence:

1. **Quality (YS I.41-43)**
   - **Being** (I.41): *Samāpatti* as the transparent crystal, Fichte's Pure Light, and Hegel's Pure Being.
   - **Existence** (I.42): *Savitarkā* as the discursive mixing (*vikalpa*), Fichte's *Anstoß*, and Hegel's Determinate Being (*Dasein*).
   - **Being-for-itself** (I.43): *Nirvitarkā* as the purified memory shining as the object, Fichte's True Infinity, and Hegel's The One.

2. **Quantity (YS I.44-46)**
   - **Quantity** (I.44): The transition to subtle objects (*sūkṣma-viṣayā*), Fichte's continuous productive imagination, and Hegel's Indifferent Limit.
   - **Quantum** (I.45): Terminating in the unmanifest (*aliṅga-paryavasānam*), Fichte's posited mathematical boundary, and Hegel's Quantum.
   - **Ratio** (I.46): The realization that this entire subtle progress is "seeded" (*sabīja*), Fichte's vicious infinite progress, and Hegel's Quantitative Ratio (Spurious Infinite).

3. **Measure (YS I.47-51)**
   - The attainment of truth-bearing insight (*ṛtambharā prajñā*). This insight acts as a Fichtean Methodological Maxim—an inhibitory mechanism (*pratibandhī*) that chokes off alien projection. Finally, perfectly matching Hegel's transition to Essence and Fichte's *Living Esse*, the mechanism itself is dropped, yielding seedless (*nirbīja*) samādhi.

This establishes the complete Fichtean/Hegelian baseline for the first chapter of the *Yoga Sūtras*.
