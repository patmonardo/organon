# Project Bootstrap: Translating the Yoga Sūtras through Dialectical Idealism

## Phase 1 Summary: The Fichtean/Hegelian Knowledge Graph

We have successfully translated key structural sequences of the *Yoga Sūtras* into rigorous Markdown-based Knowledge Graphs (the `COMPILER` workbooks) and deeply philosophical narrative essays (the `IDEA` workbooks). This dual-workbook system tracks the epistemological and ontological claims of Sāṃkhya-Yoga and maps them directly onto the transcendental architecture of J.G. Fichte's 1804 *Science of Knowing* and G.W.F. Hegel's *Science of Logic*.

### Why Move from TS-JSON to Markdown Workbooks?

The root source of our analysis remains the intricate TypeScript-JSON data structures located in `@reality/ys/`. Those TS files house the original, exhaustive analysis of Fichte's 1804 lectures as a commentary on Yoga. That system is historically rigorous but overly rigid.

We transition to the **MD Workbooks** because they are now the **Authoritative Source** for generating insight. Markdown affords us the flexibility to synthesize, refine, and iterate on these concepts discursively while preserving strict structural links (IDs, Claims, KeyPoints). The TS-JSON stands as the bedrock; the MD workbooks are the living, evolving philosophical artifacts.

### Sequences Completed (The Dialectical Ascent)
1. **Reflection (YS IV.2-11)**: The mechanics of phenomenological change and the causal "Through".
2. **Appearance (YS IV.11-21)**: The dialectic of the Thing-in-Itself and relation.
3. **Actuality (YS IV.22-34)**: Absolute Substance and the genesis of terminal liberation (*Kaivalya*).
4. **Being/Pure (YS I.41-51)**: Cognitive absorption (*Samāpatti*), truth-bearing insight (*Ṛtambharā Prajñā*), and the seedless cessation of the mirror.
5. **Concept/Absolute (YS III.53-56)**: The culmination of the Middle Way. From the atomic discrimination of time (*kṣaṇa*) to the transcendental, omniscient scope of *Tāraka*, closing the Science of Knowing entirely.

## Phase 2 Plan: Bootstrapping from the Hegelian Logic Seed

The next leap forward is to bootstrap the entire repository architecture utilizing the `@reality/ys-logic` package. 

The `ys-logic` directory contains the true Seed KG: **Hegel's Science of Logic** analyzed into a strict Knowledge Graph spanning `being`, `essence`, and `concept`.

### The Objective
Our goal is to use the Hegelian Logic KG not just as an overlay, but as the generative engine that organizes the rest of the repository. By treating Hegel's categories (Quality, Quantity, Measure, Ground, Appearance, Actuality, Subjectivity, Objectivity, Idea) as the primary coordinate system, we can:
1. Systematize the remaining Yoga Sūtras (especially the *Sādhana-pāda*).
2. Generate new, unified schemas for Abhidharma / Buddhist equivalents.
3. Drive the cypher-query architecture for the entire `@logic/compiler`.

### Structure of `@reality/ys-logic`

The Hegel Logic KG is perfectly triadic:
- `being/`: `quality/`, `quantity/`, `measure/`
- `essence/`: `reflection/`, `appearance/`, `actuality/`
- `concept/`: `subject/`, `object/`, `idea/`

Each directory contains highly structured, granular conceptual nodes, capturing Hegel's dialectical transitions.

### Bootstrapping Strategy

Instead of translating Yoga Sūtras directly into Fichte/Hegel ad hoc, we invert the process. We use the `@reality/ys-logic` KG as the **Universal Schema** for the entire `@logic/compiler`.

1. **Core Coordinate System**: The 1-3-9 structure (e.g., `essence/reflection/ground`) becomes the absolute directory structure for all comparative philosophy within the Organon.
2. **Cross-Mapping**: Future translations of Yoga, Sāṃkhya, and Abhidharma will be explicitly assigned a strict Hegelian coordinate (e.g., mapping *Sādhana-pāda* into the specific dialectic of *Measure* or *Subjectivity*).
3. **Graph Generation**: By structuring the markdown workbooks with standardized tags based on the `ys-logic` namespace, we can parse these MD files to auto-generate Cypher queries for the master Neo4j graph.

This methodology transforms the repository from a collection of translations into a unified, programmable Idea-Compiler.

### The Greater Varttika

For the Hegel analysis in the COMPILER WORKBOOKS, we have a detailed file: `reality/src/logos/ys-logic/dharma-greater-varttika.tex`. We will use this file to map the Hegel sections, breaking down Hegel's *Science of Logic* (specifically the Doctrine of Essence to start) as a direct commentary on the Yoga Sutras (Chapter 4, Dharma-pada).

**Fichte as Principle, Hegel as Varttika**: 
Fichte's *Science of Knowing* provides the overarching principles and genetic derivation of the self-constructing acts. Hegel's *Logic* serves as the **Varttika**—a complex, expansive textbook (commentary) that unpacks the dialectical movements in exhaustive categorical detail. The Hegel MD Workbooks should break down the sutra strictly following the corresponding sections matched in `dharma-greater-varttika.tex`.

New philosophical IDEAs generated for the Sanskrit analyses will be appended as follow-on essays inspired directly by this Fichte/Hegel synthesis.
