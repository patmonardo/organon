## 1) Working Thesis

The Concept as such is not a static class-term; it is a triadic operational structure:
1. **Universal** (Being as such): stable inherence, global identity frame.
2. **Particular** (Nothing as such / Shine): selective predication, weak binding.
3. **Singular** (Becoming as such): determinate this-here-now transition, strong facticity.

So `as such` names the mode in which a moment is bound to its proper logical role, not merely mentioned.

## 3) Facticity Gradient (Your Example)

- **Universal claim**: "All cats went out in the rain" — broad frame, weak immediate grounding unless instantiated.
- **Particular claim**: "Some cats are grey" — selective, factual but weakly binding; pre-determining moment.
- **Singular claim**: "This cat got wet in the rain at time t" — strongest determinable becoming, high facticity.

Interpretation: the Particular is often where meaning is *latched* but not fully compelled; it can express the unessential/contingent moment unless integrated into singular determination.

```python
from dataclasses import dataclass
from enum import Enum

class Role(str, Enum):
    UNIVERSAL = "U"
    PARTICULAR = "P"
    SINGULAR = "S"

@dataclass(frozen=True)
class Moment:
    name: str
    role: Role
    as_such: str
    facticity_strength: float  # 0..1

triad = [
    Moment("Being", Role.UNIVERSAL, "Being as such", 0.60),
    Moment("Nothing", Role.PARTICULAR, "Nothing as such (Shine)", 0.35),
    Moment("Becoming", Role.SINGULAR, "Becoming as such", 0.90),
]

for m in triad:
    print(f"{m.name:9} | role={m.role.value} | as_such={m.as_such:28} | facticity={m.facticity_strength:.2f}")
```

## 4) Implementation Hooks for Consciousness Operations

2. Role classification (`Universal`, `Particular`, `Singular`).
3. `as such` validation (is the role intrinsically fitting, or merely asserted?).
4. Facticity scoring by determinability and indexical anchoring (`this`, time, place, event).
